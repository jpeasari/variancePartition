### Merging

```python
## Loading the csv file that is refined
merge1 = pd.read_csv('ppData_immVar_02.csv')
merge1.head()
```

## GPL6244-17930.txt was obtained using the below link :
https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GPL6244
## GPL6244-17930.txt was modified and changed to a csv file that contains only two columns (ID and GE_list)
## Coulmns with no gene names are removed
## Ge_list column that has main as the name were aslo removed

```python
## Reading the GEO.csv file and changing the ID_chip column as string and dropped columns that has NAA values in it
merge2 = pd.read_csv("GEO.csv",dtype={'ID_chip': 'string'}).dropna()
merge2.head()
```

```python
## Using the left join in the merge function, I merged ppData that has ID and GEO.csv that has ID column
new_Data = merge1.merge(merge2, how='left', on='ID').dropna()
new_Data.head()
```

```python
## Next, ID column was dropped in the merged dataframe since it is no longer required
## Affymetrix probe ids were replaced with the corresponding gene names
```

```python
new_Data.drop(["ID","Unnamed: 0"], axis=1).to_csv("ppData_immVar03.csv") ## Saving the refined file to a CSV
```

```python
final_data = pd.read_csv("ppData_immVar03.csv",index_col=[0],dtype={'Gene_name': 'string'})
final_data
```

```python
## Since the Gene_name was the last column, I have made that to be the first column of the dataframe
cols = list(final_data.columns) 
cols = [cols[-1]] + cols[:-1]
final_data = final_data[cols]
final_data
```

```python
# final_data.reset_index(drop=True, inplace=True)
## As the Gene_name column has repeated gene names, I dropped columns that has duplicate gene names
## Now, total number of rows came to be 23306
final_data = final_data.drop_duplicates(subset='Gene_name', keep="last")
final_data
```

```python
## Resetting the Gene_name column as the index
final_data.set_index('Gene_name')
```

```python
## Writing to a csv file with quotes embedd to a values that are strings
import csv
final_data.to_csv("ppData_immVar_04.csv",index=False, quotechar='"', quoting=csv.QUOTE_NONNUMERIC)
```
